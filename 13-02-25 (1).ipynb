```python
#Import the Libarires
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import .csv file and convert it to a DataFrame object
df=pd.read_csv(r"C:\Users\SANJANA\Downloads\Wholesale customers data.csv")
df
```

```python
df.head()
```

```python
print(df.info())
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
X=stscaler.transform(array)
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer data')
plt.ylabel('Eucl Distance')
plt.show()
```

```python
len(set(dendo['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group=AgglomerativeClustering(n_clusters=3)
group.fit_predict(X)
```

```python
df
```

```python
cluster=group.fit_predict(X)
cluster.shape
```

```python
#model evaluation
from sklearn.metrics import silhouette_score
```

```python
# Check how well your model is making cluster
silhouette_score(X,cluster)
```

# Kmeans

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
wccs=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wccs.append(kmeans.inertia_)
```

```python
plt.plot(range(2,11),wccs)
plt.title('the elbow method')
```

```python
wccs
```

```python
group_num=pd.DataFrame(cluster,columns=['Group'])
group_num 
```

```python
plt.plot(range(2,11),wccs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wccs')
plt.show()
```

```python
model=KMeans(n_clusters=4,random_state=309)
groups=model.fit_predict(X)
```

```python
groups.shape
```

```python
groups
```

```python
group_num=pd.DataFrame(groups,columns=['Group'])
group_num  
```

```python
cust_kmeans_data=pd.concat([df,group_num],axis=1)
cust_kmeans_data
```

```python
silhouette_score(X,groups)
```
